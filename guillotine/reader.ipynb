```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1, 2, sharex="all", sharey="all", figsize=(8,4))
seaborn.heatmap(pd.read_csv("True_test.csv",index_col="Unnamed: 0")/pd.read_csv("True_train.csv",index_col="Unnamed: 0"), annot=True, ax=axs[0],cbar=False)
axs[0].set_title("With guillotine")
seaborn.heatmap(pd.read_csv("False_test.csv",index_col="Unnamed: 0")/pd.read_csv("False_train.csv",index_col="Unnamed: 0"), annot=True, ax=axs[1],cbar=False)
axs[1].set_title("Without guillotine")
plt.tight_layout()
plt.show()

fig, axs = plt.subplots(1, 1, sharex="all", sharey="all", figsize=(5,4))
seaborn.heatmap(pd.read_csv("False_test.csv",index_col="Unnamed: 0")/pd.read_csv("True_test.csv",index_col="Unnamed: 0"), annot=True, ax=axs,cbar=False)
plt.tight_layout()
plt.show()


# print(pd.read_csv("False_test.csv",index_col="Unnamed: 0"))
# plt.figure(figsize=(15,10))
# perc = 100 * ((pd.read_csv("True_test.csv",index_col="Unnamed: 0")-pd.read_csv("False_test.csv",index_col="Unnamed: 0"))/ pd.read_csv("False_test.csv",index_col="Unnamed: 0"))
# print(perc)
# seaborn.heatmap(perc, annot=True)
```
