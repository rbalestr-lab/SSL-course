```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6,4))
false_test = pd.read_csv("False_train.csv",index_col="Unnamed: 0")
falsrain = pd.concat([false_test, false_test.mean(1)], axis=1)
false_test.columns = list(false_test.columns[:-1]) + ["mean"]
seaborn.heatmap(false_test, annot=True, cbar=False,  linewidths=1, linecolor='gray')
plt.xlabel("test on")
plt.ylabel("train on")
plt.tight_layout()
plt.show()


fig = plt.figure(figsize=(6,4))
true_test = pd.read_csv("True_train.csv",index_col="Unnamed: 0")
true_test = pd.concat([true_test, true_test.mean(1)], axis=1)
true_test.columns = list(true_test.columns[:-1]) + ["mean"]
seaborn.heatmap(true_test, annot=True, cbar=False, linewidths=1, linecolor='gray')
plt.xlabel("test on")
plt.ylabel("train on")
plt.title("With guillotine")
plt.tight_layout()
plt.show()
```
